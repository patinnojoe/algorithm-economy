```python
from pathlib import Path
import sys
import pandas as pd
from glob import glob


PROJECT_ROOT = Path().resolve().parents[0]
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))


from src.config import RAW_DATA_DIR
from src.config import PROCESSED_DATA_DIR
# from src.features.transformation import build_features, process_category, process_tags
from src.features.transformation import build_features, process_category, process_tags
from src.utils.save_processed_file import save_processed
```

```python
# import dataset

files = glob(str(RAW_DATA_DIR / "youtube_raw_*.csv"))

dfs = [pd.read_csv(file) for file in files]

df = (
    pd.concat(dfs, ignore_index=True)
      .drop_duplicates(subset=["video_id"])
)

print(f"Loaded {len(files)} files")
print(f"Total unique videos: {len(df)}")

# transform dataset and save processed data
processed_df = build_features(df)
processed_df = process_tags(processed_df)
processed_df = process_category(processed_df)

# save
save_processed(processed_df)

```
